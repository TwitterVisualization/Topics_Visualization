```python
import tqdm
import numpy as np
np.set_printoptions(suppress=True)
from gensim.models import Word2Vec

w2v_path = "/mlo-container-scratch/hartley/twitter_covid_insights/insights_All/word2vec_300d.model"
model = Word2Vec.load(w2v_path)

vocab = model.wv.index_to_key
vectors = model.wv.vectors

with open("w2v_prakhar.txt","w") as f :
    for i in tqdm.tqdm(range(len(vocab)+1)) :
        if i==0 :
            f.write(f"{len(vocab)} {len(vectors[0])}\n")
        else :
            endstr = "" if i == len(vocab) else "\n"
            vector_str = " ".join(vectors[i-1].astype(str))
            f.write(f"{vocab[i-1]} {vector_str}{endstr}")
```

```python
from collections import Counter
from tqdm import tqdm_notebook
```

```python
lang_count = Counter()
tweet_counts  = []
for f in tqdm_notebook(piped_files) :
    df = pd.read_parquet(f, columns=['lang'])
    lang_count.update(df.values.flatten())
    tweet_counts.append(len(df))
lang_count.most_common()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
```

```python
def y_fmt(tick_val, pos):
    if tick_val > 1_000_000:
        val = int(int(tick_val)/1_000_000)
        return f'{val} M'
    elif tick_val > 1000:
        val = int(int(tick_val) / 1000)
        return f'{val} k'
    else:
        return tick_val
plt.bar(range(len(tweet_counts)), tweet_counts)
#plt.bar(piped_files, tweet_counts)
ax = plt.gca()
ax.yaxis.set_major_formatter(tick.FuncFormatter(y_fmt))
plt.title('distribution of nb. of tweets')
plt.xlabel('day nb')
plt.ylabel('nb of tweets')
```
